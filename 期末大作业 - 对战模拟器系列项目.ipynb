让我们从简单的任务开始-一对一的决斗。
* 您需要创建战士（类），其有两个参数 - health（等于50点）和attack（等于5点），以及一个布尔属性 - is_alive，可以是True（如果战士的health>0）。
* 此外，你必须创建第二个战斗类型 - 骑士，这应该是战士类的子类，但其参数 attack=7。
* 另外，您必须创建一个函数fight()，它将启动两个战士之间的决斗并定义其中最强的。

决斗按照以下原则进行：
- 每一个回合中，一个战士会击中另一个，被击中的战士健康的下降值将与第一个战士的攻击值相同。在那之后，第二个战士会对第一个做同样的事情。
- 如果在回合结束时，第一个战士的生命值大于0，而另一个没有，则该函数应返回true，在另一种情况下，该函数应返回false。



```python
class Warrior:
    def __init__(self):
        self.health = 50
        self.is_alive=True
        self.attack = 5
    
    @property
    def is_alive(self):
        return self.health > 0

class Knight(Warrior):
    def __init__(self):
        self.attack = 7

def fight(unit_1, unit_2):
    unit_1.helth=unit_1.health-unit_2.attack
    unit_2.helth=unit_2.health-unit_1.attack
```

当战争贩子Warlord被包括在任何一支军队中时，该军队会得到新的move_units方法。这个方法允许重新安排该军队的部队排列顺序以获得更好的战斗效果（类似田忌赛马）。重新排列不仅可以在战斗前进行，也可以在战斗中每次我方战士死亡时进行重排。重新排列的规则如下：

* 如果军队里有枪兵，他们应该摆在其他战士的面前。

* 战争贩子应该一直呆在后面，以便在需要的时候查看战斗并重新安排战士。
* 每支军队最多只能有一个战争贩子。
* 如果军队没有战争贩子，就不能使用move_units方法。
